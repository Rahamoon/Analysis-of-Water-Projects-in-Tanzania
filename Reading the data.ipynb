# Reading and checking the data

Let's start by reading the data into a pandas DataFrame.

```python
import settings
import os

import pandas as pd

data_values = pd.read_csv(os.path.join(settings.DATA_DIR, "Training set values.csv"), index_col =0)
data_labels = pd.read_csv(os.path.join(settings.DATA_DIR, "Training set labels.csv"), index_col =0)
test_values = pd.read_csv(os.path.join(settings.DATA_DIR, "Test set values.csv"), index_col =0)
```

I will then verify the shape and the first rows of each DataFrame.

```python
print data_values.shape
data_values.head()
```

```python
print data_labels.shape
data_labels.head()
```

```python
print test_values.shape
test_values.head()
```

** In order to preprocess the data , I join the training and the test sets together, so I combine them into all_data. I add an extra column 'subset', so that I can split the data later on. **

```python
data_values['subset'] = "training"
test_values['subset'] = "test"

all_data = pd.concat([data_values, test_values])
```

```python
print all_data.shape
print all_data[all_data['subset'] == "training"].shape
print all_data[all_data['subset'] == "test"].shape
all_data.iloc[[0,1,2,-3,-2,-1]]
```

```python
all_data.to_csv('processed/all_data.csv', encoding='utf-8')
```
