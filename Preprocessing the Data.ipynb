## Handling non numeric features

The data contains many categorical variables which are not numeric. In order to be able to fit them to a machine learning algorithm later on, I need to convert them to numeric lables.
To do so, I will use the [LabelEncoder](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html#sklearn.preprocessing.LabelEncoder) class from the sickit_learn library.

*LabelEncoder is a utility class to help normalize labels such that they contain only values between 0 and n_classes-1. It can also be used to transform non-numerical labels (as long as they are hashable and comparable) to numerical labels.*

Let's start by selecting the non numeric features:

```python
non_numeric_features = ['funder','installer','basin', 'public_meeting', 'ward', 'scheme_management', 'permit',
                        'extraction_type','management', 'payment_type', 'water_quality', 'quantity', 'source', 
                        'waterpoint_type', 'area']
```

```python
data[non_numeric_features].head()
```

Importing and initializing the LabelEncoder class :

```python
from sklearn.preprocessing import LabelEncoder
lab_enc = LabelEncoder()
```

```python
for feature in non_numeric_features:
    data[feature] = lab_enc.fit_transform(data[feature])
```

```python
data[non_numeric_features].head()
```

```python
training_data = data.loc[data['subset'] == 'training']
training_data = training_data.join(labels, how='inner')
test_data = data.loc[data['subset'] == 'test']
del training_data['subset']
del test_data['subset']
print training_data.shape
print test_data.shape
```
